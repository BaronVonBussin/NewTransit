```python
# Importing necessary libraries
import pandas as pd
import featuretools as ft
from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt

# Fetching the Covertype dataset
print("Downloading dataset...")
dataset = fetch_openml(data_id=1461, as_frame=True)  # Covertype dataset
data = dataset.data
data['Cover_Type'] = dataset.target  # Adding target column

print("Dataset downloaded and loaded.")

# Displaying dataset information
print("\nDataset Info:")
print(data.info())

# Creating a Featuretools EntitySet
print("\nCreating EntitySet...")
es = ft.EntitySet(id="covertype_data")

# Adding the main dataframe
es = es.add_dataframe(dataframe_name="covertype", dataframe=data, index="id", make_index=True)

# Performing Deep Feature Synthesis
print("\nGenerating features with Deep Feature Synthesis...")
feature_matrix, feature_defs = ft.dfs(
    entityset=es,
    target_dataframe_name="covertype",
    agg_primitives=["mean", "sum", "mode"],  # Aggregation primitives
    trans_primitives=["month", "year"]      # Transformation primitives
)

# Displaying the generated features
print("\nGenerated Feature Matrix (Sample):")
print(feature_matrix.head())

# Visualizing the distribution of one of the generated features
print("\nVisualizing a sample feature...")
if "sum(Cover_Type)" in feature_matrix.columns:
    plt.figure(figsize=(10, 6))
    feature_matrix["sum(Cover_Type)"].plot(kind="hist", bins=30, title="Distribution of Sum of Cover_Type")
    plt.xlabel("Sum of Cover_Type")
    plt.ylabel("Frequency")
    plt.grid()
    plt.show()

# Save a sample of the feature matrix for inspection
feature_matrix.sample(10).to_csv("feature_matrix_sample.csv")
print("Feature matrix sample saved as 'feature_matrix_sample.csv'.")
```
