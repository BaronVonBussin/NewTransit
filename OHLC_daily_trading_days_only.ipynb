```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Get SPY data
spy = yf.download('SPY', '2013-01-01', '2013-12-31')

# Create figure
fig, ax = plt.subplots(figsize=(15, 8))

# Plot OHLC bars
for i, (idx, row) in enumerate(spy.iterrows()):
    color = 'g' if row['Close'].iloc[0] >= row['Open'].iloc[0] else 'r'

    # Main bar
    ax.bar(i, row['Close'].iloc[0] - row['Open'].iloc[0],
           bottom=row['Open'].iloc[0],
           width=0.8, color=color)

    # High-low lines
    ax.plot([i, i], [row['Low'].iloc[0], row['High'].iloc[0]],
            color=color, linewidth=1)

# Configure x-axis
plt.xticks(range(0, len(spy), 5),
           [spy.index[i].strftime('%Y-%m-%d') for i in range(0, len(spy), 5)],
           rotation=45)

ax.set_title('SPY 2013 OHLC')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
